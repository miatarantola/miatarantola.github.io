```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split 

df = pd.read_csv("pollution_income_race.csv", index_col = 0)
df = df.dropna()
```

```python

data2013 = df[df["Date Local"] == 2013]

data_cities = data2013[data2013["City"] != "Not in a city"]
sns.set_theme()

sns.relplot(data = data_cities, 
           x = "Median Household Income", 
           y = "AQI Total")

```

```python
sns.relplot(data = df, 
           x = "Median Household Income", 
           y = "SO2 AQI")
```

```python
data_cities["white_alone_percent"] = data_cities["White Alone (F) %"] + data_cities["White Alone (M) %"]
data_cities["black_alone_percent"] = data_cities["Black Alone (F) %"] + data_cities["Black Alone (M) %"]

sns.relplot(data = data_cities, 
           x = "black_alone_percent", 
           y = "AQI Total")
```

```python
data_cities["white_alone_percent"] = data_cities["White Alone (F) %"] + data_cities["White Alone (M) %"]

sns.relplot(data = data_cities, 
           x = "white_alone_percent", 
           y = "AQI Total").set(title = "AQI Total v. Percentage of White Residents by County")

```

# Mapping

```python
import geopandas as gpd
import folium
import mapclassify
import matplotlib.pyplot as plt
import pandas as pd

geo_data = gpd.read_file('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/US-counties.geojson')

geo_data.id = geo_data.id.astype(str).astype(int)

#county code and state codes are the same in the two data frames

#print(data.loc[data['County Code'] == 33])
#print(geo_data.loc[geo_data['COUNTY'] == '033'])

#shifting state/county codes from objects to integers

geo_data['COUNTY'] = geo_data.COUNTY.astype(str).astype(int)
geo_data['STATE'] = geo_data.STATE.astype(str).astype(int)

merged_df = pd.merge(geo_data, df, left_on='COUNTY', right_on='County Code')
merged_df["black_alone_percent"] = merged_df["Black Alone (F) %"] + merged_df["Black Alone (M) %"]


fig, ax = plt.subplots(3)
merged_df.plot(column="AQI Total", ax=ax[0], legend=True, cmap="coolwarm")
merged_df.plot(column="Median Household Income", ax=ax[1], legend=True, cmap="warmcool")
merged_df.plot(column="black_alone_percent", ax=ax[2], legend=True, cmap="coolwarm")


```
